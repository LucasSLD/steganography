```python
import torch
from torchvision.transforms import ToPILImage
import numpy as np
import sys
sys.path.append("../iclr_17_compression/")
from test_model import plot_tensor, plot_np_array
from matplotlib.pyplot import imshow
```

```python
t_img = torch.load("../JIN_SRNet/BossBase-1.01-cover/1.pt")
t_img.shape
```

```python
%timeit np_img = t_img[0].permute(1,2,0).cpu().numpy()
```

## Conclusion
Converting torch tensors to numpy arrays is about 10x faster using permute(1,2,0).cpu().numpy() than torch to pil to numpy
