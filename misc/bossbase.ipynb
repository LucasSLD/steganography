```python
import sys
import os
import numpy as np
import matplotlib.pyplot as plt
sys.path.append("../Tools")
from my_utils import read_pgm, plot_np_array
```

```python
n = 3
bossbase_path = "../JIN_SRNet/BossBase-1.01-precover/"
files = os.listdir(bossbase_path)
img = [read_pgm(bossbase_path + file) for file in np.random.choice(files,n)]
fig, axes = plt.subplots(1,n,figsize=(15,10))
for i, ax in enumerate(axes):
    ax.tick_params(left = False, right = False , labelleft = False , labelbottom = False, bottom = False) 
    ax.imshow(img[i],cmap="gray")
    ax.set_frame_on(False)
    plt.tight_layout()
```
