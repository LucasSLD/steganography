```python
import sys
import os
import numpy as np
import matplotlib.pyplot as plt
sys.path.append("../Tools")
from my_utils import read_pgm, plot_np_array
import torch
```

```python
n = 3
bossbase_path = "../JIN_SRNet/BossBase-1.01-precover/"
cover_path    = "../JIN_SRNet/BossBase-1.01-cover/"

files = np.array(os.listdir(bossbase_path))
cover_files = np.array(os.listdir(cover_path))


index = np.random.choice(range(len(files)),n)
# print(files[index])
# print(cover_files[index])
img = [read_pgm(bossbase_path + file) for file in files[index]]
img_cover = [torch.load(cover_path + file)[0].permute(1,2,0).cpu().numpy() for file in cover_files[index]]

fig, axes = plt.subplots(1,n,figsize=(15,10))
for i, ax in enumerate(axes):
    ax.tick_params(left = False, right = False , labelleft = False , labelbottom = False, bottom = False) 
    ax.imshow(img[i],cmap="gray")
    ax.set_frame_on(False)
    plt.tight_layout()

fig, axes = plt.subplots(1,n,figsize=(15,10))
for i, ax in enumerate(axes):
    ax.tick_params(left = False, right = False , labelleft = False , labelbottom = False, bottom = False)
    ax.imshow(img_cover[i],cmap="gray")
    ax.set_frame_on(False)
    plt.tight_layout()
```
